import pandas as pd
import numpy as np
import re

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# --------------------------
# 1. Custom Engine Extractor
# --------------------------

class EngineFeatureExtractor(BaseEstimator, TransformerMixin):
    def __init__(self, column='engine'):
        self.column = column

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        engine_series = X[self.column].astype(str).fillna('')

        result = pd.DataFrame(index=X.index)
        result['has_turbo'] = engine_series.str.contains(r'\bTurbo\b', flags=re.IGNORECASE).astype(int)
        result['has_dohc'] = engine_series.str.contains(r'\bDOHC\b', flags=re.IGNORECASE).astype(int)
        result['has_sohc'] = engine_series.str.contains(r'\bSOHC\b', flags=re.IGNORECASE).astype(int)
        result['has_hybrid'] = engine_series.str.contains(r'\bHybrid\b', flags=re.IGNORECASE).astype(int)
        result['has_vtec'] = engine_series.str.contains(r'\bi-?VTEC\b', flags=re.IGNORECASE).astype(int)
        result['valves'] = engine_series.str.extract(r'(\d{2})V', expand=False).astype(float)
        result['engine_size_l'] = engine_series.str.extract(r'(\d\.\d)\s*L', expand=False).astype(float)
        layout = engine_series.str.extract(r'\b([VI]-\d)\b', expand=False).fillna('Unknown')
        result['layout'] = layout
        return result

# --------------------------
# 2. Feature Lists (example)
# --------------------------

# Assume these are your columns (adjust based on your dataset)
numerical_features = ['cylinders', 'mileage']  # example numerical columns
categorical_features = ['fuel_type', 'transmission', 'make']  # your other categorical features

# --------------------------
# 3. Preprocessing Pipelines
# --------------------------

# For standard numerical columns
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# For categorical features
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# For engine feature extraction + internal preproc
engine_pipeline = Pipeline(steps=[
    ('extractor', EngineFeatureExtractor(column='engine')),
    ('processor', ColumnTransformer([
        ('num', StandardScaler(), ['valves', 'engine_size_l']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['layout']),
    ], remainder='passthrough'))  # keep binary flags like has_turbo
])

# --------------------------
# 4. Combine All in One Pipeline
# --------------------------

full_pipeline = ColumnTransformer(transformers=[
    ('num', numerical_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features),
    ('engine', engine_pipeline, ['engine']),
])

# If needed, plug into a model:
from sklearn.ensemble import RandomForestRegressor
model_pipeline = Pipeline([
    ('preprocessor', full_pipeline),
    ('model', RandomForestRegressor(random_state=42))
])

## transmission column transformation and feature engineering 
```
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd
import re

class TransmissionFeatureExtractor(BaseEstimator, TransformerMixin):
    """
    Extract structured features from a messy 'transmission' column using regex patterns.
    """

    def __init__(self, column='transmission'):
        self.column = column

    def fit(self, X, y=None):
        return self  # stateless

    def transform(self, X):
        X = X.copy()
        col = self.column
        X[col] = X[col].astype(str).fillna('').str.lower()

        result = pd.DataFrame(index=X.index)

        # 1. Extract number of gears (e.g., 6-Speed, 7-Speed, 9 Spd)
        result['gears'] = X[col].str.extract(r'(\d{1,2})[\s-]?[sS]peed', expand=False)
        result['gears'] = pd.to_numeric(result['gears'], errors='coerce')

        # 2. Extract transmission type
        def extract_type(txt):
            if 'cvt' in txt:
                return 'CVT'
            elif 'dct' in txt or 'dual clutch' in txt:
                return 'DCT'
            elif 'manual' in txt:
                return 'Manual'
            elif 'automatic' in txt or 'a/t' in txt:
                return 'Automatic'
            elif 'tiptronic' in txt:
                return 'Automatic'  # still automatic
            elif 'battery' in txt:
                return 'Electric'
            else:
                return 'Unknown'

        result['transmission_type'] = X[col].apply(extract_type)

        # 3. Other binary flags
        result['has_tiptronic'] = X[col].str.contains(r'tiptronic', flags=re.IGNORECASE).astype(int)
        result['has_paddle'] = X[col].str.contains(r'paddle', flags=re.IGNORECASE).astype(int)
        result['has_overdrive'] = X[col].str.contains(r'overdrive', flags=re.IGNORECASE).astype(int)
        result['is_electric'] = X[col].str.contains(r'battery|electric|ev', flags=re.IGNORECASE).astype(int)

        return result
```
